```python
import pandas as pd
import numpy as np
import nltk
```

```python
df=pd.read_csv("SPAM_text.csv")
df['Class']=np.nan
df.head()
```

```python
sentences = df['Message']
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

sentences = sentences.apply(lambda text: remove_stopwords(text.lower()))
sentences.head()
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

sentences = sentences.apply(lambda text: lemmatize_words(text))
sentences.head()
```

```python
bow = {}
i = 1
for row in sentences:
  for word in row.split(" "):
    if word not in bow:
      bow[word] = i
      i += 1
```

```python
X = []
for i in sentences:
  X.append(nltk.word_tokenize(i))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.metrics import binary_focal_crossentropy
from official.nlp import optimization
```

```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
classes = {
    "ham":0,
    "spam":1,
}

# Mapping the classes
df['Class'] = df['Category'].map(classes)
```

```python
tfhub_handle_encoder = 'https://tfhub.dev/tensorflow/small_bert/bert_en_uncased_L-4_H-512_A-8/1'
tfhub_handle_preprocess = 'https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3'
```

```python
bert_model = hub.KerasLayer(tfhub_handle_encoder)
```

```python
def build_classifier_model():
  text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
  preprocessing_layer = hub.KerasLayer(tfhub_handle_preprocess, name='preprocessing')
  encoder_inputs = preprocessing_layer(text_input)
  encoder = hub.KerasLayer(tfhub_handle_encoder, trainable=True, name='BERT_encoder')
  outputs = encoder(encoder_inputs)
  net = outputs['pooled_output']
  net = tf.keras.layers.Dense(8, activation="relu")(net)
  net = tf.keras.layers.Dropout(0.1)(net)
  net = tf.keras.layers.Dense(2, activation="softmax", name='classifier')(net)
  return tf.keras.Model(text_input, net)
```

```python
classifier_model = build_classifier_model()
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metrics = tf.metrics.SparseCategoricalAccuracy()
```

```python
epochs = 5
steps_per_epoch = 5
num_train_steps = steps_per_epoch * epochs
num_warmup_steps = int(0.1*num_train_steps)

init_lr = 3e-5
optimizer = optimization.create_optimizer(init_lr=init_lr,
                                          num_train_steps=num_train_steps,
                                          num_warmup_steps=num_warmup_steps,
                                          optimizer_type='adamw')
```

```python
classifier_model.compile(optimizer=optimizer,
                         loss=loss,
                         metrics=metrics)
```

```python
from sklearn.model_selection import train_test_split

btr, bte, bltr, blte = train_test_split(sentences, df['Class'], test_size = 0.3)
```

```python
btr.shape
```

```python
btr = np.array(btr).reshape(3900,1)
```

```python
history = classifier_model.fit(x=btr, y=bltr, epochs=10)
```

```python
loss, accuracy = classifier_model.evaluate(bte, blte)

print(f'Loss: {loss}')
print(f'Accuracy: {accuracy}')
```

```python
output = classifier_model.predict(["Hi Customer, You're Pre-Qualified for a No Documentation personal loan from L&T Finance. Instant Disbursal. Apply Now! oow.pw/k0A2o6 SPIRAEA"])
```
